## Descriptive Statistics part of Exploratory Data Analysis


```python
# measures of central tendency

print(df.mean(axis=0, numeric_only=True)) #mean of whole dataset
print(df.median(axis=0, numeric_only=True)) #median of whole dataset

print(df.groupby('species').mean()) #mean of each flower's data
print(df.groupby('species').median()) #median of each flower's daya
```

```python
# trimmed means, calculated in the interquartile range
for col in df.columns.drop('species'):
    first = df[col].quantile(0.25)
    third = df[col].quantile(0.75)
    
    trimmedmean = stats.trim_mean(df[col], 0.25)
    print(f'trimmed mean for IQR for column, {col} is: {trimmedmean}')
```

```python
#measures of variability:

print(df.var(axis=0, numeric_only=True)) #variance of the whole dataset
print(df.std(numeric_only=True, axis=0)) #standard deviation of whole dataset
print(df.groupby('species').var()) # variance of each flower
print(df.groupby('species').std()) # standard deviation of each flower

#MAD calculation
mad_calc = {}
for col in df.columns.drop('species'):
    mad = (df[col] - df[col].mean()).abs().median()
    mad_calc[col] = round(mad,2)
print(f'\nThe Mean Absolute Deviation of dataset is: {mad_calc}')

#AAD calculation
aad_calc = {}
for col in df.columns.drop('species'):
    aad = (df[col] - df[col].mean()).abs().mean()
    aad_calc[col] = round(aad,2)
print(f'The Absolute Average Deviation of the dataset is: {aad_calc}')
```

```python
#skewness and kurtosis
skew = {}
kurt = {}
for col in df.columns.drop('species'):
    skewness = stats.skew(df[col],axis=0)
    skew[col] = round(skewness,2)
    kurtosis = stats.kurtosis(df[col], axis=0)
    kurt[col] = round(kurtosis,2)

print(f'The skewness of the dataset is: {skew}')
print(f'The kurtosis of the dataset is: {kurt}')
```

## Data Visualization part of Exploratory Data Analysis
## Univariate Analysis


```python
# bar charts
pl = df['petal_length'].value_counts().head(15).plot(kind='bar')
pl.set_xlabel('petal lengths')
pl.set_ylabel('frequencies')
plt.show()
```

```python
#histograms for different flowers
setosa = df[df['species']=='setosa']
setosa = setosa.drop('species',axis=1)
hist1 = setosa['sepal_length'].value_counts().plot(kind='hist') # setosa in orange
versicolor = df[df['species'] == 'versicolor']
versicolor = versicolor.drop('species',axis=1)
hist2 = versicolor['sepal_length'].value_counts().plot(kind='hist') #versicolor in blue
virginica = df[df['species'] == 'virginica']
virginica = virginica.drop('species',axis=1)
hist3 = virginica['sepal_length'].value_counts().plot(kind='hist') #versicolor in purple
plt.show()
```

```python
# box plots
sns.boxplot(data=df,x='species',y='sepal_length', hue='species').set(ylabel='Length/width for sepal in cm')
sns.boxplot(data=df,x='species',y='sepal_width', hue='species').set(ylabel='Length/width for sepal in cm')
plt.show()
sns.boxplot(data=df,x='species',y='petal_length', hue='species').set(ylabel='Length/width for petal in cm')
sns.boxplot(data=df,x='species',y='petal_width', hue='species').set(ylabel='Length/width for petal in cm')
plt.show()
```

```python
# violin plots
sns.violinplot(data=df,x='species',y='sepal_length', hue='species').set(ylabel='Length/width for sepal in cm')
sns.violinplot(data=df,x='species',y='sepal_width', hue='species').set(ylabel='Length/width for sepal in cm')
plt.show()
sns.violinplot(data=df,x='species',y='petal_length', hue='species').set(ylabel='Length/width for petal in cm')
sns.violinplot(data=df,x='species',y='petal_width', hue='species').set(ylabel='Length/width for petal in cm')
plt.show()
```

## Bivariate Analysis


```python
# scatter plots
sns.scatterplot(data=df, x='sepal_length',y='sepal_width',hue='species')
plt.show()
sns.scatterplot(data=df,x='petal_length',y='petal_width',hue='species')
plt.show()
```

```python
# contour plots
sns.kdeplot(data=df,x='petal_length',y='petal_width',cmap='Blues',fill=True)
plt.show()
sns.kdeplot(data=df,x='petal_length',y='petal_width',hue='species',fill=False)
plt.show()
```

```python
# hexagonal binning plot
sns.jointplot(data=setosa, x='petal_length', y='petal_width', kind='hex',color='b')
plt.show()
sns.jointplot(data=versicolor, x='petal_length', y='petal_width', kind='hex')
plt.show()
sns.jointplot(data=virginica, x='petal_length', y='petal_width', kind='hex')
plt.show()
sns.jointplot(data=df,x='petal_length',y='petal_width',hue='species')
plt.show()
```

## Multivariate Analysis


```python
# correlation heatmaps

cor = df.drop('species',axis=1).corr()
sns.heatmap(cor, annot=True, fmt='.3f',linewidth=.7,linecolor='pink',cmap=sns.cubehelix_palette(as_cmap=True))
plt.show()
sns.clustermap(cor, annot=True, fmt='.3f',linewidth=.7,linecolor='pink',cmap=sns.cubehelix_palette(as_cmap=True))
plt.show()
```
